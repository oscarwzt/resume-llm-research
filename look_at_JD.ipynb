```python
def clean_text(text):
    #text = text.replace('\n', ' ')
    #text = text.replace('\t', ' ')
    text = text.replace('\r', ' ')
    text = text.replace('  ', ' ')
    return text

def format_job_description(row):
    position = row['position']
    company = row['company']
    description = row['description']
    description = clean_text(description)
    return f'Here is a job description for the role {position} at {company}:\n\n"{description}"'

def get_query_for_gpt4(num_resumes=3, keys = keys, json_format = True, formatted_resume = False, raw_resume = False):
    selected = np.random.choice(keys, num_resumes, replace=False)
    for key in selected:
        print("Resume id:", key)
        print()
        if json_format:
            print(json.dumps(resumes[key], indent=4))
            print()
        elif formatted_resume:
            print(create_resume_from_json(resumes[key]).replace("- ? "," "))
        elif raw_resume:
            print(getCleanResume(resumes[key]).replace("- ? "," "))
        print()
```
